# ***Standarization***

```python
originaltext=text.lower() #turing all text to lower case
```

# ***Tokenization***

```python
import re
sentences=(re.split(r'[.,!?]', originaltext))
sentences=list(dict.fromkeys(sentences)) #removing redundant data on the text
```

```python
text=' '.join([str(sentence) for sentence in sentences]) #converting the sentences to a bulk of text for the next processing
text
```

```python
words= word_tokenize(text)
stop_words =stopwords.words("english")
freqTable = dict()
punctuation= '''!()-[].{};:'"\, <>./?@#$%^&*_~'''
```

# ***Stemming***

```python
def stemTokens(wordsFiltered):
  from nltk.stem import PorterStemmer
  from nltk.stem import LancasterStemmer
  #create an object of class PorterStemmer
  porter = PorterStemmer()
  lancaster=LancasterStemmer()
  stemmedWords=[]
  for wordFiltered in wordsFiltered:
   
    stemmedWord=porter.stem(wordFiltered)
    stemmedWords.append(stemmedWord)
  return stemmedWords
```

```python
words=stemTokens(words)
words
```

# ***Lemmatiation***

```python
def lemmTokens(stemmedWords):
  from nltk.stem import WordNetLemmatizer
  wordnet_lemmatizer = WordNetLemmatizer()
  lemmitizedWords=[]
  for stemmedWord in stemmedWords:
    lemmitizedWord=wordnet_lemmatizer.lemmatize(stemmedWord)
    lemmitizedWords.append(lemmitizedWord)
  return(lemmitizedWords)
```

```python
words=lemmTokens(words)
words
```

```python
type(words)
```

# ***Words Frequency Table***

```python
for word in words:
  if word in (stop_words or punctuation):
     continue
  if word in freqTable:
     freqTable[word]+=1
  else: 
    freqTable[word]=1
```

```python
freqTable.items()
```

# ***text to Sentence Spliiting***

```python
# sentece tokenizer
import re
sentences=(re.split(r'[.,!?]', originaltext))
sentenceValue=dict()
sentences
```

# ***Sentence Score Table***

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
porter = PorterStemmer()
```

```python
for sentence in sentences:
  terms=word_tokenize(sentence)
  for term in terms:
   for word in freqTable :
    rootedterm=porter.stem(term)
    rootedterm=wordnet_lemmatizer.lemmatize(rootedterm)
    #print(term,rootedterm, word)
    if word == rootedterm:
      if sentence in sentenceValue:
        sentenceValue[sentence]+=freqTable[word]
      else:
        sentenceValue[sentence]=freqTable[word]
```

```python
sentenceValue
```

```python
len(sentenceValue)
```

```python
allSentencesSum=0
for sentence in sentenceValue:
  allSentencesSum+=sentenceValue[sentence]
```

```python
allSentencesAverage=int(allSentencesSum/len(sentenceValue))
allSentencesAverage
```

```python
allSentencesSum
```

# ***Evaluating Score***

```python
summary=''
for sentence in sentences:
  if (sentence in sentenceValue) and (sentenceValue[sentence]>(allSentencesAverage)):
    summary+= " "+sentence

print(summary)
```

```python
with open("summarizedText.txt", "w") as f:
    
  f.writelines(summary)
        
```
